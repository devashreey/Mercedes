```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Import Dataset

data = pd.read_csv('train.csv')
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
data.head()
```

```python
pd.options.display.float_format='{:,.3f}'.format
variance=data.var(numeric_only=True)
variance=variance.reset_index()
variance.columns=['id','values']
variance
```

```python
variance=variance.sort_values('values',ascending=1)
variance
```

```python
var=variance.loc[variance['values']<0.1,'id']
data1=data.drop(var,axis=1)
```

```python
data1.drop('ID',axis=1,inplace=True)
data1.head()
```

```python
# checking for null values:

data1[data1.isnull().any(axis=1)]
```

```python
# check for unique records:

data1.nunique()
```

```python
# remove duplicate records:

data1[data1.duplicated()==True]
```

```python
data1.duplicated().sum()
```

```python
data1=data1.drop_duplicates()
```

```python
# create seperate objects to store categorical and numeric variables:

char=data1.select_dtypes(exclude='number')
num=data1.select_dtypes(include='number')
char
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
#data1['X0']=label_encoder.fit_transform(data1['X0'])
```

```python
# Label Encoder convert categorical columns into numerical ones 

for col in char.columns:
    label_encoder=LabelEncoder()
    data1[col]=label_encoder.fit_transform(data1[col])
    data1[col].unique()

data1.shape
```

```python
# Scale the data:

from sklearn import preprocessing
data1_scaled=preprocessing.scale(data1)
```

```python
# perform dimentionality reduction:
# Transform variables store all the data
from sklearn.decomposition import PCA
pca=PCA(n_components=2).fit(data1_scaled)
```

```python
data_pca=pca.transform(data1)
data_pca.shape
```

```python
# xgboost model:

X=data1.drop('y',axis=1)
y=data1.loc[:,'y']
```

```python
# splitting:

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=123)
```

```python
import xgboost as xg
```

```python
# seed and random_state both are same

xgb_reg=xg.XGBRegressor(objective='reg:squarederror',n_estimators=10,seed=123)
```

```python
xgb_reg.fit(X_train,y_train)
```

```python
# predict:
y_pred=xgb_reg.predict(X_test)
```

```python
d=pd.DataFrame()
d['y_test']=y_test
d['y_pred']=y_pred
d['mape']=abs((d['y_test']-d['y_pred'])/d['y_test']) # MAPE - Mean Absolute Percentage Error
```

```python
(d.mape.mean())*100
```
